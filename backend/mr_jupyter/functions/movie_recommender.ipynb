```python
def convert3(obj):
    L =[]
    counter = 0
    for i in ast.literal_eval(obj):
      if counter!=3:
        L.append(i['name'])
        counter+=1
      else:
          break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
def fetchcrew(obj):
    L =[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetchcrew)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'] 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(enumerate(similarity[movie_index]),reverse=True,key=lambda x:x[1])[1:6]

    for i in distances:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
